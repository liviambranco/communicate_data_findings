## Investigation Overview

In this investigation, I wanted to look at patterns of behavior of users of the Bay Wheels bikeshare system. 

## Dataset Overview


Bay Wheels is a regional public bicycle sharing system in the San Francisco Bay Area, California operated by Motivate in a partnership with the Metropolitan Transportation Commission and the Bay Area Air Quality Management District. Bay Wheels is the first regional and large-scale bicycle sharing system deployed in California and on the West Coast of the United States. It was established as Bay Area Bike Share in August 2013. As of January 2018, the Bay Wheels system had over 2,600 bicycles in 262 stations across San Francisco, East Bay and San Jose. On June 28, 2017, the system officially re-launched as Ford GoBike in a partnership with Ford Motor Company. After Motivate's acquisition by Lyft, the system was subsequently renamed to Bay Wheels in June 2019. The system is expected to expand to 7,000 bicycles around 540 stations in San Francisco, Oakland, Berkeley, Emeryville, and San Jose.
Source: Wikipedia (https://en.wikipedia.org/wiki/Bay_Wheels)

This dataset contains information from the start of the month of June to the start of the month of July 2020. It has 22 columns and 79900 entries with information regarding bike trips. The columns include information regarding the bike rentals such as the type of bike rented, when and where the rental started and ended, how long the ride was and information about the user type. 

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
trip_data = pd.read_csv('trip_data_clean.csv')
```

## Type of Users

In June, most of the users were casual, representing a percentage of approx. 54% of all the users.

```python
color_base = sns.color_palette()[0]
graph = sns.catplot(data=trip_data, x='member_casual', kind = 'count', color = color_base, height =6);
graph.set_axis_labels('User Type', 'Monthly Rentals');
graph.fig.suptitle('User Types and Monthly Rental', fontsize=20, fontweight='bold', y=1.05);
```

## Bike Types

In June, both groups of users used docked bikes more often than electric bikes.

```python
plt.figure(figsize=[10, 8])
sns.countplot(data=trip_data, x='member_casual', hue='rideable_type', color=color_base)
plt.xlabel('Bike Type', fontsize='x-large');
plt.ylabel('Monthly Rentals',  fontsize='x-large');
plt.legend (labels =('electric','docked'), loc = 2)
plt.title('Bike Types and Monthly Usage', fontsize=20, fontweight='bold', y=1.05);
```

# Users activity

In june, during the early hours of the day, members are most active, probably going to work or school. While after 10 o'clock  casual users become more active.
Casual and members have a balanced usage of the system during the weekdays, while on weekends casual users use the system more often.

```python
plt.figure(figsize=[12, 8]);
sns.countplot(data=trip_data, x='start_hour', hue='member_casual', color = color_base);
plt.xlabel('Hour of the Day', fontsize='x-large');
plt.ylabel('Monthly Rentals', fontsize='x-large');
plt.title('User Activity per Hour', fontsize=20, fontweight='bold', y=1.05);
plt.legend (loc = 2);
```

```python
weekday = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

graph = sns.catplot(data=trip_data, x='day_of_week', col='member_casual', kind='count', sharey = True, color = color_base, order = weekday);
graph.set_axis_labels('Day of the Week', 'Number of Bike Trips');
graph.set_titles('{col_name}');
graph.set_xticklabels(rotation=45);
graph.fig.suptitle('User Activity per Day', fontsize=20, fontweight='bold', y=1.05);
```

## Average Daily Trip Duration

In June, casual users often rode for longer periods of time than members.

```python
weekday = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.figure(figsize=[10, 8])
sns.pointplot(data=trip_data, x='day_of_week', y='duration_minutes', hue='member_casual', dodge=0.3, linestyles='', color=color_base, order = weekday);
plt.xlabel('Day of Week', fontsize='x-large');
plt.ylabel('Avg. Trip Duration in Minutes', fontsize='x-large');
plt.legend (loc = 0)
plt.title('Average Trip Duration per User Type', fontsize=20, fontweight='bold', y=1.05);
plt.xticks(rotation=45);
```

## Average Daily Trip Distance

In June, casual users often rode for longer distances than members.

```python
weekday = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.figure(figsize=[10, 8])
sns.pointplot(data=trip_data, x='day_of_week', y='ride_distance_km', hue='member_casual', dodge=0.3, linestyles='', color=color_base, order = weekday);
plt.xlabel('Day of Week', fontsize='x-large');
plt.ylabel('Avg. Trip Distance (KM)', fontsize='x-large');
plt.legend (loc = 2);
plt.title('Average Trip Distance per User Type', fontsize=20, fontweight='bold', y=1.05);
plt.xticks(rotation=45);
```
